```python
#model building on sentimental

import numpy as np
import pandas as pd
```

```python
ds=pd.read_csv('/content/drive/MyDrive/data sets/spam.csv',encoding='unicode_escape')
```

```python
ds.head()
```

```python
ds.isnull().sum()
```

```python
ds.shape
```

```python
for i in ds['Category']:
  ds['Category'].replace({'ham':1,'spam':0},inplace=True)
```

```python
ds.head()
```

```python
#data cleaning

import re
import nltk
```

```python
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

all_stopwords = stopwords.words('english')
all_stopwords.remove('not')
```

```python
corpus=[]

for i in range(0, 5570):
  review = re.sub('[^a-zA-Z]', ' ', ds['Message'][i])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus[:5]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
x_fresh=cv.fit_transform(corpus)
```

```python
x_fresh[:5]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_fresh,ds['Category'],test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
model.score(x_test,y_pred)
```
